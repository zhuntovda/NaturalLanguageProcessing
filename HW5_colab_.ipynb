```python
train_data = []
for sent in full_train:
    form_opus_list = [] 
    for token in sent:
        #print(token.form, token.upos)
        form_opus_list.append((token.form, token.upos))
    train_data.append(form_opus_list)

test_data = []
for sent in full_test:
    form_opus_list = []
    for token in sent:
      form_opus_list.append((token.form, token.upos))   
    test_data.append(form_opus_list)

print(len(train_data))
```

```python
test_sent = [x[0] for x in train_data[1]]
```

```python
from nltk.tag import UnigramTagger, BigramTagger, TrigramTagger
```

```python
unigram_tagger = UnigramTagger(train_data)
display(unigram_tagger.tag(test_sent), unigram_tagger.evaluate(test_data))
```

```python
bigram_tagger = BigramTagger(train_data, backoff=unigram_tagger)
display(bigram_tagger.tag(test_sent), bigram_tagger.evaluate(test_data))
```

```python
trigram_tagger = TrigramTagger(train_data, backoff=bigram_tagger)
display(trigram_tagger.tag(test_sent), trigram_tagger.evaluate(test_data))
```

```python
from nltk.tag import DefaultTagger
```

```python
def backoff_tagger(train_sents, tagger_classes, backoff=None):
    for cls in tagger_classes:
        backoff = cls(train_sents, backoff=backoff)
    return backoff


backoff = DefaultTagger('NOUN') 
tag = backoff_tagger(train_data,  
                     [UnigramTagger, BigramTagger, TrigramTagger],  
                     backoff = backoff) 
  
tag.evaluate(test_data) 
```

```python
import tensorflow as tf

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Embedding, GlobalAveragePooling1D, Conv1D, GRU, LSTM, Dropout, GlobalMaxPooling1D
from tensorflow.keras.layers.experimental.preprocessing import TextVectorization

from razdel import tokenize
from sklearn import model_selection, preprocessing, linear_model

from sklearn.feature_extraction.text import CountVectorizer
```

```python
dir = 'Collection5/'
records = load_ne5(dir)

word_docs = []
types_docs = []

for ix, rec in enumerate(records):
  words = []
  types = []
  for token in tokenize(rec.text):
    type_ent = 'OUT'
    for ent in rec.spans:
      if (token.start >= ent.start) and (token.stop <= ent.stop):
        type_ent = ent.type
        break
    words.append(token.text)
    types.append(type_ent)
  word_docs.extend(words)
  types_docs.extend(types)

print(f'{word_docs[10]}  {types_docs[10]}')
```

```python
set(types_docs)
```

```python
train_x, valid_x, train_y, valid_y = model_selection.train_test_split(word_docs, types_docs)

# labelEncode целевую переменную
encoder = preprocessing.LabelEncoder()
train_y = encoder.fit_transform(train_y)
valid_y = encoder.fit_transform(valid_y)

count_vect = CountVectorizer(analyzer='word', token_pattern=r'\w{1,}')
count_vect.fit(word_docs)

xtrain_count =  count_vect.transform(train_x)
xvalid_count =  count_vect.transform(valid_x)
```

```python
train_data = tf.data.Dataset.from_tensor_slices((train_x, train_y))
valid_data = tf.data.Dataset.from_tensor_slices((valid_x, valid_y))

train_data = train_data.batch(16)
valid_data = valid_data.batch(16)
```

```python
def custom_standardization(input_data): 
    return input_data

vocab_size = 30000
seq_len = 120

vectorize_layer = TextVectorization(
    standardize=custom_standardization,
    max_tokens=vocab_size,
    output_mode='int',
    output_sequence_length=seq_len)

# Make a text-only dataset (no labels) and call adapt to build the vocabulary.
text_data = train_data.map(lambda x, y: x)
vectorize_layer.adapt(text_data)
```

```python
embedding_dim = 64

class model_NER(tf.keras.Model):

  def __init__(self):
    super(model_NER,self).__init__()
    self.emb = Embedding(vocab_size,embedding_dim)
    self.conv1 = Conv1D(300, (6))
    self.conv2 = Conv1D(300, (6))
    self.gPool = GlobalMaxPooling1D()
    self.fc1 = Dense(300,activation='relu')
    self.fc2 = Dense(50,activation='relu')
    self.fc3 = Dense(6,activation='softmax')

  def call(self, x):
    
    x = vectorize_layer(x)
    x = self.emb(x)

    x = self.conv1(x)
    x = self.conv2(x)

    pool_x = self.gPool(x)

    fc_x = self.fc1(pool_x)
    fc_x = self.fc2(fc_x)

    concat_x = tf.concat([pool_x, fc_x], axis=1)
    prob = self.fc3(fc_x)

    return prob
```

```python
model_NER = model_NER()
```

```python
model_NER.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(),
              metrics=['accuracy'])
```

```python
model_NER.fit(train_data, validation_data=valid_data, epochs=3)
```
