1. Удалим @user из всех твитов с помощью паттерна "@[\w]*". Для этого создадим функцию: 
 - для того, чтобы найти все вхождения паттерна в тексте, необходимо использовать re.findall(pattern, input_txt)
 - для для замены @user на пробел, необходимо использовать re.sub()


```python
def findall_and_sub_df(df,pattern,col='tweet'):
    
    for index, row in df.iterrows():
        if len(re.findall(pattern, row[col])) != 0:
            df.at[index, col] = re.sub(pattern, ' ', row[col])
    
    return df
```

```python
combine_df = findall_and_sub_df(combine_df, '@[\w]*')
combine_df.head()
```

2. Изменим регистр твитов на нижний с помощью .lower()

```python
def lower_df(df,col='tweet'):
    
    for index, row in df.iterrows():
        df.at[index,col] = row[col].lower()
    
    return df
```

```python
combine_df = lower_df(combine_df)
combine_df.head()
```

3. Заменим сокращения с апострофами (пример: ain't, can't) на пробел, используя apostrophe_dict. Для этого необходимо сделать функцию: для каждого слова в тексте проверить (for word in text.split()), если слово есть в словаре apostrophe_dict в качестве ключа (сокращенного слова), то заменить ключ на значение (полную версию слова).

```python
def sub_df_in_mask(df,dict_sub,string_punctuation=',.',col='tweet'):
    
    for index, row in df.iterrows():
        for key in ''.join([ch for ch in list(row[col]) if ch not in string_punctuation]).split():
            if key in dict_sub.keys(): 
                df.at[index, col] = re.sub(key, dict_sub[key], row[col])
                #print(key)
    return df
    
```

```python
combine_df = sub_df_in_mask(combine_df,apostrophe_dict,string.punctuation[:6] + string.punctuation[7:])
combine_df.head()
```

4. Заменим сокращения на их полные формы, используя short_word_dict. Для этого воспользуемся функцией, используемой в предыдущем пункте.

```python
combine_df = sub_df_in_mask(combine_df,short_word_dict,string.punctuation[:14] + string.punctuation[15:])
combine_df.head()
```

9. Удалим из текста слова длиной в 1 символ, используя ' '.join([w for w in x.split() if len(w)>1])

```python
def del_bit_words_df(df,word_len=1,col='tweet'):
    
    for index, row in df.iterrows():
        df.at[index, col] = ' '.join([w for w in row['tweet'].split() if len(w)>word_len])
        
    return df
```

```python
combine_df = del_bit_words_df(combine_df)
combine_df.head()
```

10. Поделим твиты на токены с помощью nltk.tokenize.word_tokenize, создав новый столбец 'tweet_token'.

```python
from nltk import tokenize as tknz
from nltk.corpus import stopwords
```

```python
def creat_word_tokenize(df,col='tweet',col_token='tweet_token'):
    
    df[col_token] = ''
    
    for index, row in df.iterrows():
        df.at[index, col_token] = tknz.word_tokenize(row[col])
        
    return df
```

```python
combine_df = creat_word_tokenize(combine_df)
combine_df.head()
```

11. Удалим стоп-слова из токенов, используя nltk.corpus.stopwords. Создадим столбец 'tweet_token_filtered' без стоп-слов.

```python
def del_word_tokenize(df,col_token='tweet_token',col_token_filt='tweet_token_filtered'):
    
    df[col_token_filt] = ''
    stop_words = stopwords.words("english")
    
    for index, row in df.iterrows():
        df.at[index, col_token_filt] = [word for word in row[col_token] if not word in stop_words]
        
    return df
```

```python
combine_df = del_word_tokenize(combine_df)
combine_df.head()
```

12. Применим стемминг к токенам с помощью nltk.stem.PorterStemmer. Создадим столбец 'tweet_stemmed' после применения стемминга.

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
def stemmer_df(df,col='tweet_token_filtered',col_stemmed='tweet_stemmed'):
    
    df[col_stemmed] = ''
    stemmer = PorterStemmer()
    
    for index, row in df.iterrows():
        for i in range(len(row[col])):
            row[col][i] = stemmer.stem(row[col][i])
        df.at[index, col_stemmed] = row[col]
        
    return df
```

```python
combine_df = stemmer_df(combine_df)
combine_df.head()
```

13. Применим лемматизацию к токенам с помощью nltk.stem.wordnet.WordNetLemmatizer. Создадим столбец 'tweet_lemmatized' после применения лемматизации.

```python
def lemmatized_df(df,col='tweet_token_filtered',col_lemmatized='tweet_lemmatized'):
    
    df[col_lemmatized] = ''
    stemmer = PorterStemmer()
    
    for index, row in df.iterrows():
        for i in range(len(row[col])):
            row[col][i] = stemmer.stem(row[col][i])
        df.at[index, col_lemmatized] = row[col]
        
    return df
```

```python
combine_df = lemmatized_df(combine_df)
combine_df.head()
```
